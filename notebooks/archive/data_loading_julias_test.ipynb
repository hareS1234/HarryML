```python
import pandas as pd
```

```python
data = pd.read_csv("data/data extraction/ALA_SER/ALA_SER_FUS_sequential.csv")
data
```

```python
data.columns
```

```python
new_columns = []
last_column_name = None
for column in data.columns:
    if column.startswith("Unnamed"):
        new_columns.append(last_column_name+"_Y")
    else:
        last_column_name = column
        new_columns.append(last_column_name+"_X")
        
```

```python
new_columns
```

```python
data = data.drop(0)
data
```

```python
gene_data = {}
buf_column = ""
for column in data.columns:
    if column.endswith("X"):
        if buf_column:
            gene_data[(column, buf_column)] = []
            buf_column = ""
        else:
            buf_column = column
    elif column.endswith("Y"):
        if buf_column:
            gene_data[(buf_column, column)] = []
            buf_column = ""
        else:
             buf_column = column
```

```python
gene_data
```

```python
for v in zip(data["WT_X"], data["WT_Y"]):
    print(v)
```

```python
for gene_name in gene_data:
    #print(gene_name)
    for v in zip(data[gene_name[0]], data[gene_name[1]]):
        print(v)
        if not pd.isna(v[0]) and not pd.isna(v[1]):
            gene_data[gene_name].append(v)
gene_data
```

```python
for k, v in gene_data.items():
    gene_data[k] = sorted(v)
```
