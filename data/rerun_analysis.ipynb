```python
errors = []
simulation_root_dir = "FINISHED SIMULATIONS"
for seq_name in next(os.walk(simulation_root_dir))[1]:
    
    if not os.path.exists(os.path.join(simulation_root_dir, seq_name, "ps_densities")):
        os.mkdir(os.path.join(simulation_root_dir, seq_name, "ps_densities"))

    for file in next(os.walk(os.path.join(simulation_root_dir, seq_name, "ps_densities")))[2]:
         os.remove(os.path.join(simulation_root_dir, seq_name, "ps_densities", file))
    
        
    shutil.copy(os.path.join(SIMULATION_LIBS, "ppd.py"), os.path.join(simulation_root_dir,
                                                                      seq_name,
                                                                      "ps_densities"))

   
    for temp in next(os.walk(os.path.join(simulation_root_dir, seq_name)))[1]:
        if temp != "ps_densities" and temp != '.ipynb_checkpoints':
           
            shutil.copy(os.path.join(SIMULATION_LIBS, "da.py"), (os.path.join(simulation_root_dir, seq_name, temp, "da.py")))
            print(os.path.join(simulation_root_dir, seq_name, temp))
          
            r = subprocess.run(["python", "da.py", "densities_chunked2.dat"],
                                   cwd=os.path.join(simulation_root_dir, seq_name, temp),
                                   capture_output=True)
            if r.returncode != 0:
                errors.append(r)
                continue
            if temp.startswith("temp"):
                temp_v = temp[4:]
            else:
                temp_v = temp
            shutil.copy(os.path.join(simulation_root_dir, seq_name, temp, "densities_chunked2.dat_concentrations.txt"),
                      os.path.join(simulation_root_dir, seq_name, "ps_densities", f"densities_chunked_temp{temp_v}.txt"))
        
    shutil.copy(os.path.join(SIMULATION_LIBS, "ppd.py"), (os.path.join(simulation_root_dir, seq_name, "ps_densities", "ppd.py")))
    subprocess.run(["python", "ppd.py"], cwd=os.path.join(simulation_root_dir, seq_name, "ps_densities"))
```
